Import python modules

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np    
```

```python
def calculatemothly(model):
    monthyear = pd.read_csv('dataset/monthyear.csv') 
    feature = pd.read_csv('outputData/ricePrediction/'+model+'_rice_waterUsage.csv') 
    df = pd.concat( [monthyear, feature], axis=1 )
    df.drop(['Unnamed: 0'],axis=1,inplace=True)
    df = df.groupby(['Year','Month']).sum()
    #df.to_csv('outputData/mothly/'+model+'_rice_monthly.csv')
```

calculate monthly water usage for each of models

```python
calculatemothly('linear')
calculatemothly('knn')
calculatemothly('kernal')
calculatemothly('dt')
```

```python
def plotMonthlyusage(prediction,name):
    label = pd.read_csv('outputData\monthly/label_rice_monthly.csv')
    xlabel = pd.read_csv('dataset/x.csv')
    print('Label: ',label.shape)
    label = np.array(label)
    label = label.reshape((-1))
    xlabel = np.array(xlabel)
    prediction = np.array(prediction)

    plt.figure(figsize=(12,8))
    names = xlabel
    x = range(len(names))
    y = label
    y1=prediction
    plt.xlim(-1, 90) 
    plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'label line')
    plt.plot(x, y1, marker='*', ms=10,label=u'prediction line')
    plt.legend() 
    plt.margins(0)
    plt.subplots_adjust(bottom=0.15)
    plt.xlabel('Months') 
    plt.ylabel("Monthly Water Usage") 
    plt.title(name)
    plt.show()
```

```python
kernal = pd.read_csv('outputData/monthly/kernal_rice_monthly.csv')
plotMonthlyusage(kernal,'Kernal Model')
```

```python
knn = pd.read_csv('outputData/monthly/KNN_rice_monthly.csv')
plotMonthlyusage(knn,'KNN Model')
```

```python
DT = pd.read_csv('outputData/monthly/DT_rice_monthly.csv')
plotMonthlyusage(DT,'Decision Tree Model')
```

```python
linear = pd.read_csv('outputData/monthly/linear_rice_monthly.csv')
plotMonthlyusage(linear,'Linear Model')
```
